Dieses Skript visualisiert die Power Spektren nach dem Filtern und Resamplen, um zu überprüfen, dass alles korrekt funktoiniert hat.

```python
import pyedflib
import matplotlib.pyplot as plt
from scipy.signal import welch

original_path = '/home/users/jhischke/tuar_processed_backup_before_resampling/test/aaaaaall_s004_t000.edf'
new_path = '/home/users/jhischke/tuar_processed/test/aaaaaall_s004_t000.edf'
channel = 'F7-T3'

with pyedflib.EdfReader(original_path) as reader:
    channel_idx = reader.getSignalLabels().index(channel)
    channel_data = reader.readSignal(channel_idx)
    sampling_frequency = reader.getSampleFrequency(channel_idx)
    f, pxx = welch(channel_data, fs=sampling_frequency, nperseg=2048)
    plt.semilogy(f, pxx, label='Original', color='blue')

with pyedflib.EdfReader(new_path) as reader:
    channel_idx = reader.getSignalLabels().index(channel)
    channel_data = reader.readSignal(channel_idx)
    sampling_frequency = reader.getSampleFrequency(channel_idx)
    f, pxx = welch(channel_data, fs=sampling_frequency, nperseg=2048)
    plt.semilogy(f, pxx, label='After resampling/filter', color='orange')

plt.title(f"Power Spektrum des EEG Signals von Kanal {channel}")
plt.xlabel("Frequenz in Hz")
plt.ylabel("Leistung/Frequenz (V²/Hz)")
plt.legend()
plt.savefig('./output/exp_02_07_power_spectrum_comparison_1.svg')
plt.show()
```
