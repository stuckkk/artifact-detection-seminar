Dieses Skript visualisiert zum einen den Anteil der Patienten, die einen epileptischen Schock erlitten haben und zum Anderen, wie viel Prozent der Aufnahmen einen epileptischen Schock aufweisen.

```python
import os
import matplotlib.pyplot as plt

total_patients = set()
seizure_patients = set()

total_files = 0
seizure_files = 0


for root, dirs, files in os.walk('../../../../tuar/edf'):
    for file in files:
            if file.endswith('.edf'):
                total_files += 1
                total_patients.add(file.split('_')[0])
            if file.endswith('_seiz.csv'):
                seizure_files += 1
                seizure_patients.add(file.split('_')[0])
```

```python
plt.title(f"Anteil der Dateien mit einem epileptischen Schock (n={total_files})")
plt.pie([seizure_files, total_files - seizure_files], labels=['Schock', 'Kein Schock'], autopct='%1.1f%%')
plt.savefig('output/exp_01_08_seizure_file_distribution.svg')
plt.show()
```

- Notizen:
    - Knapp 14% der Dateien beinhalten einen epileptischen Schock. Das ist mehr als ich gedacht hätte. Allerdings werden EEG Aufnahmen wahrscheinlich von zu Anfällen neigenden Patienten gemacht.

```python
plt.title(f"Anteil der Patienten mit einem epileptischen Schock (n={len(total_patients)})")
plt.pie([len(seizure_patients), len(total_patients) - len(seizure_patients)], labels=['Schock', 'Kein Schock'], autopct='%1.1f%%')
plt.savefig('output/exp_01_08_seizure_patient_distribution.svg')
plt.show()
```

- Notizen
    - Der Anteil ist ziemlich ähnlich zu dem Anteil der epileptischen Dateien.
