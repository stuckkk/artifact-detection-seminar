Dieses Skript visualisiert die Verteilungen der Features.

```python
from utils.visualization import visualize_feature_distribution
import h5py

set = 'train'
data_split_path = 'data_split.yaml'
hdf5_path = 'features/features.hdf5'

with h5py.File(hdf5_path) as f:
    # Get any of the channels to list all the features
    session_group = f[next(iter(f))]
    channel_group = session_group[next(iter(session_group))]

    features = [feat for feat in channel_group if not feat == 'labels']

for feature in features:
    figs, axes = visualize_feature_distribution(feature, hdf5_path, set, data_split_path)
    for i, fig in enumerate(figs):
        fig.savefig(f'output/exp_03_05_{feature}{f'_{i+1}' if len(figs) > 1 else ''}_distribution.svg')
```
