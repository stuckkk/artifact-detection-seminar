Dieses Skript erzeugt die Labels für die gegebenen Daten und speichert diese im HDF5 Format.

```python
from utils.persistence import save_session_labels
from utils.labeling import label_all_files

window_size_sec = 1.0
window_overlap = 0.0
overlap_treshold = 0.3
dir_path = "../../../../tuar_processed"
hdf5_file_path = f"./labelled_data/labels.hdf5"

label_generator = label_all_files(dir_path, window_size_sec, window_overlap, overlap_treshold)

for session, label_dict in label_generator:
    save_session_labels(hdf5_file_path, session, label_dict)
```
